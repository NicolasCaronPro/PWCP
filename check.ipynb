```python
fireDataset = dataset[dataset['isFire'] > 0]
test = fireDataset[fireDataset['cluster'] == 9]
test[['creneau', 'isFire', 'day_since_fire']]
```

```python
dataset[dataset['isFire'] > 0]
```

```python
import numpy as np
from datetime import datetime
day_since_fires = np.ones(len(test))
ind = 0
### Get days since last fire
for _, row in test[1:].iterrows():
    dff = test[test['creneau'] < row['creneau']]
    if len(dff) == 0:
        val = np.nan
        ind += 1
        continue
    
    val = (datetime.strptime(row['creneau'], '%Y-%m-%d') - (datetime.strptime(dff[-1:]['creneau'].values[0], '%Y-%m-%d'))).days
    day_since_fires[ind] = val
    ind += 1
```
